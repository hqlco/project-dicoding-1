```python
datetime_3 = ["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"]

for column in datetime_3:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

Menambah Delivery Time

```python
delivery_time = orders_df["order_delivered_customer_date"] - orders_df["order_delivered_carrier_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

## Exploratory Data Analysis (EDA)

### customers_df

Melihat jumlah pelanggan berdasarkan customer_city

```python
customers_df.groupby(by="customer_city").customer_unique_id.nunique().sort_values(ascending=False).reset_index()
```

Melihat jumlah pelanggan berdasarkan customer_state

```python
customers_df.groupby(by="customer_state").customer_unique_id.nunique().sort_values(ascending=False).reset_index()
```

menghitung jumlah nilai unik dari order_id untuk setiap payment_type

```python
order_payments_df.groupby(by="payment_type").order_id.nunique().sort_values(ascending=False).reset_index()
```

Merge customers_df dan orders_df

```python
merge1_df = pd.merge(
    left=customers_df,
    right=orders_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
merge1_df.head()
```

Merge order_payments dan order_reviews

```python
merge2_df = pd.merge(
    left=order_payments_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
merge2_df.head()
```

```python
merge2_df.groupby(by="payment_type").agg({
    "order_id": "nunique",
    "payment_value":  ["min", "max", "mean"]
})
```

Merge merge1_df & merge2_df

```python
main1_df = pd.merge(
    left=merge1_df,
    right=merge2_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
main1_df.head()
```

### Explore order_items_df dan sellers_df

Merge order_items_df dan sellers_df

```python
merge3_df = pd.merge(
    left=order_items_df,
    right=sellers_df,
    how="left",
    left_on="seller_id",
    right_on="seller_id"
)
merge3_df.head()
```

```python
merge3_df.groupby(by="seller_city").seller_id.nunique().sort_values(ascending=False).reset_index()
```

### Explore products_df dan  product_category_name_translation_df

Merge products_df dan product_category_name_translation_df

```python
merge4_df = pd.merge(
    left=products_df,
    right=product_category_name_translation_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
merge4_df.head()
```

```python
merge4_df.groupby(by="product_category_name_english").product_id.nunique().sort_values(ascending=False).reset_index()
```

Merge merge4 dan merge3

```python
main2_df = pd.merge(
    left=merge4_df,
    right=merge3_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
main2_df.head()
     
```

```python
main2_df.groupby(by="product_category_name_english").agg({
    "order_id": "nunique",
    "price":  ["min", "max", "mean"]
})
```

Explore Semua Data

```python
main_data = pd.merge(
    left=main1_df,
    right=main2_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
main_data.head()
```

```python
main_data.info()
```

```python
main_data.groupby(by=["product_category_name_english", "customer_city"]).agg({
    "price": "sum",
    "freight_value": "sum"
}).sort_values(by=["price", "freight_value"], ascending=False)
```

```python
main_data.groupby(by="customer_state").agg({
    "order_id": "nunique",
    "payment_value": "sum"
}).sort_values(by="payment_value", ascending=False)
     
```

Convert main_data to .csv

```python
main_data.to_csv('main_data.csv', index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa yang paling laris dan paling sepi terjual?

```python
item_data1 = main_data.groupby("product_category_name_english")["product_id"].count().reset_index()
item_data1 = item_data1.sort_values(by="product_id", ascending=False)
item_data1.head()
     
```

```python
item_data2 = main_data.groupby("product_category_name_english")["product_id"].count().reset_index()
item_data2 = item_data2.sort_values(by="product_id", ascending=True)
item_data2.head()
```

```python
# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(x="product_id", y="product_category_name_english", data=item_data1.head(10), color="skyblue", legend=False)

# Menambahkan label dan judul
plt.xlabel("Jumlah Produk")
plt.ylabel("Kategori Produk (Bahasa Inggris)")
plt.title("Jumlah Produk Paling Laris")

# Menampilkan plot
plt.show()
     
```

```python

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(x="product_id", y="product_category_name_english", data=item_data2.head(10), color="skyblue", legend=False)
# Menambahkan label dan judul
plt.xlabel("Jumlah Produk")
plt.ylabel("Kategori Produk (Bahasa Inggris)")
plt.title("Jumlah Produk Paling Sepi")

# Menampilkan plot
plt.show()
```

### Pertanyaan 2: Bagaimana penyebaran dari penjual berdasarkan kota?

```python
sellers_in_cities = main_data.groupby(by="seller_city").seller_id.nunique().sort_values(ascending=False).reset_index()
sellers_in_cities.head()
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x="seller_id", y="seller_city", data=sellers_in_cities.head(10),  color="skyblue", legend=False)


plt.xlabel("Jumlah Penjual")
plt.ylabel("Kota")
plt.title("penyebaran dari penjual berdasarkan kota")


plt.show()
```
